```python
import pandas
import matplotlib.pyplot as plt
import pylab
```

```python
data = pandas.read_excel("../data/后验差比表.xlsx")
data
```

```python
p = data["plot"]
SOC = data["SOC"]
SIC = data['SIC']
N = data['N']
g = data['good']
q = data['qualified']
b = data['bare']
```

```python
#plt.rcParams["figure.dpi"] = 300
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(4, 2), dpi=300) 
ax = data.plot.scatter(x='plot',y='SOC',marker = 'o',color='DarkBlue',label = "SOC")
bx = data.plot.scatter(x='plot',y='SIC',marker = '+',color='Red',ax=ax,label = "SIC")
cx = data.plot.scatter(x='plot',y='N',ax=bx,marker = 'x',color='green',label = "N")
dx = data.plot(x = 'plot',y='good',ax = cx,color='grey',linestyle = 'dashed',label = "好")
ex = data.plot(x = 'plot',y='qualified',ax = dx,color='orange',linestyle = 'dashed',label = "合格")

fx = data.plot(x = 'plot',y='bare',ax = ex,color='red',linestyle = 'dashed',label = "勉强")

plt.legend(prop={"family": "SimHei", "size": 8},fancybox=True,shadow=True,loc='upper right')
plt.xticks(fontname="Times New Roman", fontsize=10)
plt.yticks(fontname="Times New Roman", fontsize=10)
plt.xlabel('区块',fontsize=12)
plt.ylabel('后验差比值',fontsize=12)
```
