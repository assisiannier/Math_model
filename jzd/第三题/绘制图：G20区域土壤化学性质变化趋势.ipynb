```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.read_excel("../data/碳氮检测数据集/G20-HGI.xlsx")
data_mean=data.groupby('year').mean()
data_mean

data_mean
```

```python
data_mean.to_csv("G9-HGI.csv")
```

```python
import pylab
plt.figure(figsize=(3, 2), dpi=300) 
soc=data_mean["SOC土壤有机碳"]
sic=data_mean["SIC土壤无机碳"]
n=data_mean["全氮N"]
soc.plot()
sic.plot()
n.plot()
pylab.xticks(rotation=45)
plt.xticks(fontname="Times New Roman", fontsize=6)
plt.yticks(fontname="Times New Roman", fontsize=6)
plt.xlabel('time',fontname="Times New Roman",fontsize=8)
plt.ylabel('g',fontname="Times New Roman",fontsize=8)
plt.legend(["SOC", "SIC","N"],prop={"family": "Times New Roman", "size": 6},fancybox=True,shadow=True,loc='upper right')
```

```python
import matplotlib.ticker as ticker
import pylab
plt.figure(figsize=(3,2), dpi=300) 
CDN = data_mean["土壤C/N比"]
CDN.plot()
pylab.xticks(rotation=45)
plt.xticks(fontname="Times New Roman", fontsize=6)
plt.yticks(fontname="Times New Roman", fontsize=6)
plt.xlabel('time',fontname="Times New Roman",fontsize=8)
plt.ylabel('%',fontname="Times New Roman",fontsize=8)
plt.legend(["C/N"],prop={"family": "Times New Roman", "size": 6},fancybox=True,shadow=True,loc='upper right')
```
